```python
def sortstring(x):
    t=list(x)
    t=sorted(t)
    return ''.join(t)
```

```python
def all_anagram(x):
    d=dict()
    for line in open(x):
        words=line.strip().lower()
        s=sortstring(words)
        if s not in d:
            d[s]=[words]
        else:
            d[s].append(words)
    return d
```

```python
def all_anagram_sorted(x):
    d=dict()
    l=[]
    for line in open(x):
        words=line.strip().lower()
        s=sortstring(words)
        if s not in d:
            d[s]=[words]
        else:
            d[s].append(words)
    for j in d:
        l.append((len(d[j]),d[j],j))
    l=list(reversed(sorted(l)))
    return l
```

```python
def bingo(x):
    d=all_anagram_sorted(x)
    l=[]
    b=[]
    for i,j,k in d:
        if len(k)==8:
            l.append(i)
    m=max(l)
    for i,j,k in d:
        if len(j)==m and len(k)==8:
            b.append(j)
    return b
   
    
```

```python
def count(x,y):
    count=0
    if len(x) != len(y):
        print("length is different")
    else:
        for i in range(len(x)):
            if x[i]!=y[i]:
                count+=1
    return count
```

```python
def metathesis_list(l):
    x=[]
    for i in range(len(l)):
        for j in range(len(l)):
            if count(l[i],l[j])==2:
                x.append((l[i],l[j]))
    return x
```

```python
def metathesis(x):
    d=all_anagram(x)
    l=[]
    x=[]
    for i in d:
        l.append(metathesis_list(d[i]))
    for i in l:
        if len(i)>0:
            x.append(i)
    return x
            
```
